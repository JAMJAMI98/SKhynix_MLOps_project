# 방법1) 전체 df에 moving average 적용

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import TimeSeriesSplit, train_test_split

# 1) 윈도우 사이즈에 따른 이동 평균을 계산하는 함수
def calculate_moving_average(df, window_size, target_column):
    """
    주어진 윈도우 사이즈로 이동 평균을 계산하고,
    타겟 컬럼과 비교하여 MSE 계산
    """
    # 이동 평균
    ma_column = df[target_column].rolling(window=window_size, min_periods=1).mean()

    # 한 스텝씩 옮겨서 비교...MSE 계산을 위해
    y_true = df[target_column].iloc[window_size+1:].to_numpy()
    y_pred = ma_column.iloc[window_size:-1].to_numpy()  # 한 스텝 앞으로

    mse = mean_squared_error(y_true, y_pred)
    return mse

# 윈도우 사이즈 범위는 일단 1부터 50..나중에 바꿔보기
window_sizes = range(1, 51)

# 타겟 컬럼 지정
target_column = 'max_val_1'

# 각 윈도우 사이즈에 대한 MSE 저장 딕셔너리
mse_scores = {}

for window_size in window_sizes:
    mse = calculate_moving_average(df, window_size, target_column)
    mse_scores[window_size] = mse
    print(f"Window Size: {window_size}, MSE: {mse}")

# 최적의 윈도우 사이즈 출력
best_window_size = min(mse_scores, key=mse_scores.get)
print(f"Best Window Size: {best_window_size}")
```

```python
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np

def calculate_moving_average(df2, window_size, target_column):
    try:
        ma_column = df2[target_column].rolling(window=window_size, min_periods=1).mean()
        y_true = df2[target_column].iloc[window_size+1:].to_numpy()
        y_pred = ma_column.iloc[window_size:-1].to_numpy()
        mse = mean_squared_error(y_true, y_pred)
        return mse
    except ValueError as e:
        if 'Input contains NaN' in str(e):
            return None  # NaN이 포함되어 있으면 None 반환
        else:
            raise  # 다른 종류의 오류는 그대로

value_types = ['max', 'mean', 'median', 'min', 'p05', 'p10', 'p90', 'p95', 'range', 'slope', 'stddev']
prmt_ids = range(1, 20)

window_sizes = range(1, 51)
best_window_sizes = {}

for value_type in value_types:
    for prmt_id in prmt_ids:
        target_column = f"{value_type}_val_{prmt_id}"
        mse_scores = {}
        for window_size in window_sizes:
            mse = calculate_moving_average(df2, window_size, target_column)
            if mse is not None:  # MSE 계산 결과가 유효한 경우에만 저장
                mse_scores[window_size] = mse
        if mse_scores:  # 유효한 MSE 점수가 있을 경우에만 최적의 윈도우 사이즈 계산
            best_window_size = min(mse_scores, key=mse_scores.get)
            best_window_sizes[target_column] = best_window_size
            print(f"{target_column}: Best Window Size = {best_window_size}, MSE = {mse_scores[best_window_size]}")
        else:
            print(f"{target_column}: skip_NaN")
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import TimeSeriesSplit, train_test_split

def calculate_moving_average(df, window_size, target_column):

    ma_column = df[target_column].rolling(window=window_size, min_periods=1).mean()

    # 한 스텝씩 옮겨서 비교...MSE 계산을 위해
    y_true = df[target_column].iloc[window_size+1:].to_numpy()
    y_pred = ma_column.iloc[window_size:-1].to_numpy()  # 한 스텝 앞으로

    mse = mean_squared_error(y_true, y_pred)
    return mse

# 윈도우 사이즈 범위 설정
window_sizes = range(1, 51)

# 결과 저장을 위한 딕셔너리
results = {}

# df['STEP']의 모든 유니크한 값에 대해 반복
for step in df['STEP_NM'].unique():
    step_df = df[df['STEP_NM'] == step]  # STEP 별 데이터 필터링

    # 각 윈도우 사이즈에 대해 MSE 계산
    for window_size in window_sizes:
        mse = calculate_moving_average(step_df, window_size, target_column)

        # 결과 저장
        if step not in results:
            results[step] = {}
        results[step][window_size] = mse

# 결과 출력
for step, mses in results.items():
    best_window_size = min(mses, key=mses.get)
    best_mse = mses[best_window_size]
    print(f"STEP {step}: Best Window Size = {best_window_size}, Lowest MSE = {best_mse}")
```
